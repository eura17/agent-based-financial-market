```python
import pandas as pd
```

```python
df = pd.read_csv("../data/DP_LIVE_23042023160716976.csv")
```

```python
quarterly = df.loc[
    (df["LOCATION"] == "USA")
    & (df["INDICATOR"] == "QGDP")
    & (df["SUBJECT"] == "TOT")
    & (df["FREQUENCY"] == "Q"),
    "Value",
] / 100
daily = (1 + quarterly) ** (1 / 90) - 1
```

```python
daily.describe()
```

```python
quarterly.quantile(0.95)
```

```python
quarterly.plot.hist(
    bins=60,
    grid=True,
    figsize=(10, 6),
);
```

```python
quarterly.loc[(-0.06 <= quarterly) & (quarterly <= 0.06)].describe()
```

```python
daily.loc[(-0.06 <= quarterly) & (quarterly <= 0.06)].plot.hist(
    bins=60,
    grid=True,
    figsize=(10, 6),
);
```
